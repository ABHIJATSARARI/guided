```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 
print('Modules are imported.')
```

import dataset

```python
corona_dataset_csv = pd.read_csv("covid19_Confirmed_dataset.csv")
corona_dataset_csv.head(10)
```

```python
corona_dataset_csv.shape
```

delete useless columns 

```python
df = corona_dataset_csv.drop(["Lat","Long"],axis=1)
```

```python
df.head(10)
```

Aggregating row by country name 

```python
corona_dataset_aggregate=df.groupby("Country/Region").sum()
corona_dataset_aggregate.head()
```

```python
corona_dataset_aggregate.shape
```

Visualization data related to a country

```python
corona_dataset_aggregate.loc["China"][:3].plot()
# corona_dataset_aggregate.loc["Italy"].plot()
# corona_dataset_aggregate.loc["S"]
```

calculating and ploting the first derivative of the curve

```python
corona_dataset_aggregate.loc["China"].diff().plot()
```

find the maximum infection rate for china , italy and spain

```python
corona_dataset_aggregate.loc["China"].diff().max()
```

```python
corona_dataset_aggregate.loc["Italy"].diff().max()
```

```python
corona_dataset_aggregate.loc["Spain"].diff().max()
```

find the maximum infection rate of all country 

```python
countries = list(corona_dataset_aggregate.index)
max_infection_rates = []
for c in countries:
    max_infection_rates.append(corona_dataset_aggregate.loc[c].diff().max())
corona_dataset_aggregate["max_infection_rate"]=max_infection_rates
corona_dataset_aggregate.head()
```

create a new dataframe with needed columns

```python
corona_data = pd.DataFrame(corona_dataset_aggregate["max_infection_rate"])
corona_data.head()
```

importing the data set 

```python
happiness_report_csv = pd.read_csv("worldwide_happiness_report.csv")
happiness_report_csv.head(10)
```

```python
useless_cols = ["Overall rank","Score","Generosity","Perceptions of corruption"]
```

```python
happiness_report_csv.drop(useless_cols,axis=1,inplace=True)
happiness_report_csv.head()
```

charging the indices of the dataframe

```python
happiness_report_csv.set_index("Country or region",inplace=True)
happiness_report_csv.head()
```

```python
data = corona_data.join(happiness_report_csv,how="inner")
data.head()
```
